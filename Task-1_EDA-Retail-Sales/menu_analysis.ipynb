## Working on the Second Dataset to find meaningful insights and give recommendation by performing EDA.

## 1. Data Loading and Cleaning: Load the retail sales dataset.

```python
# Importing Specific library which are required for data pre-processing tasks.

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Loading the dataset and checking the first 5 rows of the dataset.

df = pd.read_csv('menu.csv')
df.head(5)
```

```python
# Making a copy of the dataset to perform data pre-processing tasks.

df_copy = df.copy()
```

```python
# Checking the shape, non-null values, and Statistical summary( for numerical data) of the dataset.

print("Shape of the dataset:", df_copy.shape)
print("\nNon-null values in the dataset:", df_copy.info())
display("\nStatistical summary of the dataset:", df_copy.describe())
```

- Here we can see that the dataset contains 260 rows and 24 columns. The dataset has no null values, and the data types of the columns are appropriate for analysis. It is a proper web scrapped dataset of McDonald's menu items, which includes various attributes such as item name, category, price, calories, and other nutritional information. The dataset is ready for further analysis and exploration to derive insights about the menu items and their characteristics.

```python
# Checking the missing values in the dataset.

print("Checking for missing values in the dataset:")
print(df_copy.isnull().sum())
```

## 2. Descriptive Statistics: Calculate basic statistics (mean, median, mode, standard deviation).

```python
numerical_cls = df_copy.select_dtypes(include=[np.number])
print("\nMean of numerical columns:\n", numerical_cls.mean())
print("\nMedian of numerical columns:\n", numerical_cls.median())
print("\nMode of numerical columns:\n", numerical_cls.mode().iloc[0])
print("\nStandard deviation of numerical columns:\n", numerical_cls.std())
```

## 3. Customer and Product Analysis: Category-wise Nutritional Breakdown

```python
# Number of items per category.

category_counts = df_copy['Category'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='Set2', hue=category_counts.index)
plt.title('Number of Menu Items per Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- As we can see from the Barplot that the 'Coffee & Tea' category is highly taken by the customers. The 'Breakfast' category is the second most popular category among the customers.

```python
# Average calories per category
avg_calories = df_copy.groupby('Category')['Calories'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=avg_calories.index, y=avg_calories.values, palette='OrRd', hue=avg_calories.index, legend=False)
plt.title('Average Calories per Category')
plt.xlabel('Category')
plt.ylabel('Avg Calories')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- As we can see from the barplot that the 'Chicken & Fish' category has the highest average calories among all the categories, followed by 'Burgers'. The 'Coffee & Tea' category has the lowest average calories, which is expected as it mainly consists of beverages. This analysis can help customers make informed choices based on their dietary preferences and calorie intake goals.

## 4. Top 10 Highest Calorie Items

```python
top10_calories = df_copy[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x='Calories', y='Item', data=top10_calories, palette='Reds_r', hue='Item', legend=False)
plt.title('Top 10 Highest Calorie Items')
plt.tight_layout()
plt.show()
```

- As we can see from the horizontal barplot that the 'Chicken McNuggets (40 pieces)' is the highest calorie item on the menu, followed by 'Big Breakfast with Hotcakes'. This information can be useful for customers who are looking to manage their calorie intake and make healthier choices when ordering from the menu.

## 5. Visualization: Fat vs Calories by Category

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_copy, x='Total Fat', y='Calories', hue='Category', s=80)
plt.title('Total Fat vs Calories by Category')
plt.xlabel('Total Fat (g)')
plt.ylabel('Calories')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

- From the scatter plot, we can observe a positive correlation between total fat and calories across different categories. Items in the 'Chicken & Fish' and 'Burgers' categories tend to have higher total fat and calorie content compared to items in the 'Coffee & Tea' category, which generally have lower values for both. This visualization can help customers understand the nutritional content of different menu items and make informed decisions based on their dietary preferences and health goals.

## 6. Correlation Heatmap of Nutritional Features

```python
key_cols = ['Calories', 'Total Fat', 'Saturated Fat', 'Sugars', 'Protein', 'Sodium', 'Carbohydrates']
corr = df_copy[key_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='seismic', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Key Nutritional Features')
plt.tight_layout()
plt.show()
```

- From the correlation heatmap, we can see that the Calories and the Total Fat, Saturated Fat, and Protein are strongly positively correlated, indicating that as the calorie content of an item increases, its fat and protein content also tends to increase. On the other hand, there is a negative correlation between Sugar and Sodium, protein are strongly negatively correlated, suggesting that items with higher sugar content tend to have lower sodium and protein levels. These insights can help customers make informed choices based on their nutritional preferences and dietary goals when selecting menu items.

## 7. Recommendations

Based on the Exploratory Data Analysis performed on McDonald's Menu Nutrition Dataset:

1. **Reduce High-Calorie Items in Chicken & Fish Category**
   The 'Chicken & Fish' category has the highest average calories among all categories,
   followed by 'Burgers'. McDonald's should introduce grilled or baked alternatives
   to reduce the overall calorie load in these categories.

2. **Address the Top 10 Calorie Offenders**
   Items like 'Chicken McNuggets (40 pieces)' and 'Big Breakfast with Hotcakes' are
   the highest calorie items on the menu. Offering smaller portion sizes or
   lighter ingredient alternatives for these items would help health-conscious customers.

3. **Leverage the Popularity of Coffee & Tea**
   The 'Coffee & Tea' category has the highest number of menu items and the lowest
   average calories. McDonald's should continue expanding this category with
   low-sugar, low-calorie beverages to attract health-focused customers.

4. **Control Fat Content to Directly Reduce Calories**
   The scatter plot and heatmap both confirm a strong positive correlation between
   Total Fat and Calories. Reducing fat content in Burgers and Chicken items
   will directly bring down their calorie count — the most efficient improvement strategy.

5. **Introduce Low-Sugar Alternatives in High-Sugar Categories**
   The heatmap shows Sugar and Sodium are negatively correlated — meaning high-sugar
   items (like shakes and desserts) tend to be lower in sodium but still unhealthy
   due to sugar content. Introducing naturally sweetened or sugar-free variants
   would significantly improve the menu's nutritional profile.

6. **Better Labelling for Informed Customer Choices**
   Since Calories, Total Fat, and Saturated Fat are strongly correlated, McDonald's
   should clearly display these three metrics together on menu boards and the app
   so customers can make quick, informed decisions without needing to read
   all 24 nutritional columns.
